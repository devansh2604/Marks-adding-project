```python
import csv

# Open the CSV file
f = open("marks.csv", "r")
csvreader = csv.reader(f)

# Read and print the header
header = next(csvreader)
print(header)

# Read all remaining rows into a list
rows = list(csvreader)    #csvrader never in set , and list gives list of list

# Print each row
for row in rows:
    print(row)

# Create dictionary for total marks
totalmarks = {}

# Accumulate total marks per student
for row in rows:

    if row[0] in totalmarks:
        totalmarks[row[0]] += row[1]
    else:
        totalmarks[row[0]] = row[1]

# Print total marks
print(totalmarks)
# Close the file
f.close()
```
